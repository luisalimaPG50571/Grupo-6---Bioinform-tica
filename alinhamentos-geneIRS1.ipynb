```python
from Bio.Blast import NCBIXML
from Bio.Blast import NCBIWWW
from Bio import SeqIO
```

```python
seq_IRS1= SeqIO.read(open("IRS1.fasta"), format="fasta")
print (len(seq_IRS1.seq))
```

Devido ao grande comprimento da sequência nucleotídica do gene IRS1, o programa não retornava o resultado do qblast. Sendo assim, guardamos o resultado do blast (https://blast.ncbi.nlm.nih.gov/Blast.cgi#) num ficheiro .xml e carregamos esse ficheiro para análise em Biopyhton 

```python
#result_handle = NCBIWWW.qblast("blastn", "nr", seq_IRS1.format("fasta"))
result_handle = open("sequenciasIRS1.xml")
record = NCBIXML.read(result_handle)
```

```python
print("Parâmetros")
print("Database:" , record.database)
print("Matrix:", record.matrix)
print("Gap penalties:", record.gap_penalties)
```

```python
e_value_threshold = 0.0000001
c = len(seq_IRS1.seq)

print("Sequências homólogas:")
print("")
for alignment in record.alignments:
    for hsp in alignment.hsps:
        n=hsp.align_length
        i=(hsp.identities/n)
        q=(n/c)
        p=(hsp.positives/n)
        if hsp.expect < e_value_threshold:
            print("Acesso:" , alignment.accession)
            print("Hit id:" , alignment.hit_id)
            print("E-value:" , hsp.expect)
            print("Comprimento:", hsp.align_length)
            print("Identidade:" , (i*100), "%")
            print("Query cover:" , (q*100), "%")
            print("Positivies:", (p*100), "%")
            print("Definição:" ,alignment.hit_def)
            print("---Alinhamento----", "Query start:", hsp.query_start, "Sbjct start:",hsp.sbjct_start)
            print("")
        
```
